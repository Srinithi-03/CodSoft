```python
import pandas as pd
df=pd.read_csv('/content/Titanic-Dataset.csv')
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
val=df.drop('Survived',axis='columns')
to_predict=df.Survived
d=pd.get_dummies(val.Sex)
val=pd.concat([val,d],axis='columns')
val.drop('Sex',axis='columns',inplace=True)
val.head()
```

```python
to_predict.head()
```

```python
val.columns[val.isna().any()]
```

```python
val.Age=val.Age.fillna(val.Age.mean())
val.columns[val.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
a,b,c,d=train_test_split(val,to_predict,test_size=0.2)
m=GaussianNB()
m.fit(a,c)
```

```python
m.score(b,d)
```
